```python
from google.cloud.sql.connector import Connector, IPTypes
import sqlalchemy
from sqlalchemy.sql import text
import os
import pandas as pd
```

```python
# admin key
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "python-api-key\cloudtest-400719-dce1ea0134cc.json" 
```

```python
# initialize Cloud SQL Connector
connector = Connector()

# SQLAlchemy database connection creator function
def getconn():
    conn = connector.connect(
        "cloudtest-400719:us-west1:test-cloud-sql", # Cloud SQL Instance Connection Name
        "pymysql",
        user="root",
        password="1245",
        db="test",
        #ip_type=IPTypes.PUBLIC # IPTypes.PRIVATE for private IP
    )
    return conn

# create SQLAlchemy connection pool
pool = sqlalchemy.create_engine(
    "mysql+pymysql://",
    creator=getconn,
)

# interact with Cloud SQL database using connection pool
with pool.connect() as db_conn:

    # query database
    result = db_conn.execute(text('select * from testable'))

    # Do something with the results
    df = pd.DataFrame(result.fetchall())

# close Cloud SQL Connector

connector.close()
```
